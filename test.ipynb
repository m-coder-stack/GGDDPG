```python
from link_tree import build_tree, draw_map
import numpy as np
from grid_world import calculate_link_speed, calculate_link_speed_center

# np.random.seed(0)
uav_num = 20
center_position = np.zeros((1,2))
uav_position = np.concatenate((np.random.uniform(-1000,1000,(uav_num,1)), np.random.uniform(-100,100,(uav_num,1))), axis = 1)
client_position = np.random.uniform(-1000,100,(10,2))
uav_traffic = np.random.uniform(5,10,(uav_num))
print(uav_traffic.sum())

center_node_speed_list = calculate_link_speed_center(uav_position, center_position)
uav_speed_matrix = calculate_link_speed(uav_position)
center_node  = build_tree(center_position, center_node_speed_list, uav_position, uav_speed_matrix, uav_traffic)
tree = center_node.build_tree()
print(center_node.traffic_load)
print(tree.show(stdout=False))
draw_map(tree, center_position, uav_position, client_position, size = [[-1000,1000], [-1000,1000]])

```

```python
import matplotlib.pyplot as plt
import numpy as np
from solver import solve_param

# 定义 x 变量的范围 (-3，3) 数量 50 
x_range = 1000
x=np.linspace(-x_range,x_range,200)

param_B = 20
param_alpha = solve_param(100, 100)

print(param_alpha)


y = param_B * np.log2((1 + param_alpha / x**2).astype(np.float32))

y = np.clip(y, 0, 50)

# Figure 并指定大小
plt.figure(num=3,figsize=(8,5))
# 绘制 y=x^2 的图像，设置 color 为 red，线宽度是 1，线的样式是 --
plt.plot(x,y,color='red',linewidth=1.0,linestyle='--')

# 设置 x，y 轴的范围以及 label 标注
plt.xlim(-x_range,x_range)
plt.ylim(0,100)
plt.xlabel('x')
plt.ylabel('y')

# 设置坐标轴刻度线 
# Tick X 范围 (-1，2) Tick Label(-1，-0.25，0.5，1.25，2) 刻度数量 5 个
# new_ticks=np.linspace(-1,2,5)
# plt.xticks(new_ticks)

# Tick Y 范围(-2.2,-1,1,1.5,2.4) ，Tick Label (-2.2, -1, 1, 1.5, 2.4) 别名(下面的英文)
# plt.yticks([-2.2,-1,1,1.5,2.4],
#           [r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])


# 设置坐标轴 gca() 获取坐标轴信息
ax=plt.gca()
# 使用.spines设置边框：x轴；将右边颜色设置为 none。
# 使用.set_position设置边框位置：y=0的位置；（位置所有属性：outward，axes，data）
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 移动坐标轴
# 将 bottom 即是 x 坐标轴设置到 y=0 的位置。
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

# 将 left 即是 y 坐标轴设置到 x=0 的位置。
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


# 显示图像
plt.show()
print(y[0])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一些数据点
x = np.random.rand(50)
y = np.random.rand(50)

# 生成一个与数据点相对应的值，这些值将用于颜色映射
values = np.linspace(0, 1.5, 50)

sizes = np.linspace(10, 100, 50)

# 使用scatter函数绘制点，c参数指定颜色数据，cmap指定颜色映射
plt.scatter(x, y, c=values, s=sizes, cmap='coolwarm')

# 添加颜色条
plt.colorbar(label='Value scale')

# 添加标题和标签
plt.title('Points with Red to Blue Gradient')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# 显示图形
plt.show()
```
